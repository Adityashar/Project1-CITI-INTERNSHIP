```python
lis_word = []
for row in rq_data:
    for word in row:
        lis_word.append(word)
        
for row in aq_data:
    for word in row:
        lis_word.append(word)
        

lis_word = list(set(lis_word))
```

```python
print(lis_word)
```

```python
word_2_indices = {val:index for index, val in enumerate(lis_word)}
indices_2_word = {index:val for index, val in enumerate(lis_word)}
```

```python
with open("word2index.pkl", "wb") as p:
    pickle.dump(word_2_indices,p)
```

```python
with open("index2word.pkl", "wb") as p:
    pickle.dump(indices_2_word, p)
```
